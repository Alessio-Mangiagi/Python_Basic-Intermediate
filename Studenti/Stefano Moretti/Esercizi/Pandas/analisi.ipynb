Distribuzione delle statistiche per tutti in pokémon

```python
df["HP Stat"].hist(color='orange', edgecolor='black')
plt.xlabel("HP Stat")
plt.ylabel("Frequency")
plt.title("Distribution of HP Stats for all Pokémon")
plt.show()

df["Attack Stat"].hist(color='orange', edgecolor='black')
plt.xlabel("Attack Stat")
plt.ylabel("Frequency")
plt.title("Distribution of Attack Stats for all Pokémon")
plt.show()

df["Special Attack Stat"].hist(color='orange', edgecolor='black')
plt.xlabel("Special Attack Stat")
plt.ylabel("Frequency")
plt.title("Distribution of Special Attack Stats for all Pokémon")
plt.show()

df["Defense Stat"].hist(color='orange', edgecolor='black')
plt.xlabel("Defense Stat")
plt.ylabel("Frequency")
plt.title("Distribution of Defense Stats for all Pokémon")
plt.show()

df["Special Defense Stat"].hist(color='orange', edgecolor='black')
plt.xlabel("Special Defense Stat")
plt.ylabel("Frequency")
plt.title("Distribution of Special Defense Stats for all Pokémon")
plt.show()

df["Speed Stat"].hist(color='orange', edgecolor='black')
plt.xlabel("Speed Stat")
plt.ylabel("Frequency")
plt.title("Distribution of Speed Stats for all Pokémon")
plt.show()

```
