```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('pokemon1_5gen.csv')
```

```python
df.head(10)
```

```python
fig = plt.figure(figsize=(18, 12)) # Dimensione della figura
fig.patch.set_facecolor('lightblue')  # Colore di sfondo della figura

fig.suptitle('Frequenza delle Statistiche di tutti i PokÃ©mon dalla 1Âª alla 5Âª Generazione', fontsize=20, fontweight='bold')

# Statistiche totali (sopra, tutta la riga)
ax1 = plt.subplot2grid((3, 3), (0, 1))
ax1.hist(df['Base Stat Total'], bins=20, color='paleturquoise', edgecolor='black')
ax1.set_title('Statistiche totali') 
ax1.set_xlabel('Totale')
ax1.set_ylabel('Frequenza')
ax1.grid(axis='y', alpha=0.75)

# PS
ax2 = plt.subplot2grid((3, 3), (1, 0))
ax2.hist(df['Health Stat'], bins=20, color='lightpink', edgecolor='black')
ax2.set_title('Punti Salute (PS)')
ax2.set_xlabel('PS')
ax2.set_ylabel('Frequenza')
ax2.grid(axis='y', alpha=0.75)

# Attacco
ax3 = plt.subplot2grid((3, 3), (1, 1))
ax3.hist(df['Attack Stat'], bins=20, color='saddlebrown', edgecolor='black')
ax3.set_title('Attacco')
ax3.set_xlabel('Attacco')
ax3.set_ylabel('Frequenza')
ax3.grid(axis='y', alpha=0.75)
ax3.set_xlim(right=250)  
ax3.set_ylim(top=160)  # Imposta il limite superiore dell'asse y

# Difesa
ax4 = plt.subplot2grid((3, 3), (1, 2))
ax4.hist(df['Defense Stat'], bins=20, color='slategrey', edgecolor='black')
ax4.set_title('Difesa')
ax4.set_xlabel('Difesa')
ax4.set_ylabel('Frequenza')
ax4.grid(axis='y', alpha=0.75)
ax4.set_xlim(right=250)  
ax4.set_ylim(top=160)

# Attacco Speciale
ax5 = plt.subplot2grid((3, 3), (2, 0))
ax5.hist(df['Special Attack Stat'], bins=20, color='mediumorchid', edgecolor='black')
ax5.set_title("Attacco Speciale")
ax5.set_xlabel('Attacco Speciale')
ax5.set_ylabel('Frequenza')
ax5.grid(axis='y', alpha=0.75)
ax5.set_xlim(right=250)  
ax5.set_ylim(top=160)

# Difesa Speciale
ax6 = plt.subplot2grid((3, 3), (2, 1))
ax6.hist(df['Special Defense Stat'], bins=20, color='tan', edgecolor='black')
ax6.set_title("Difesa Speciale")
ax6.set_xlabel('Difesa Speciale')
ax6.set_ylabel('Frequenza')
ax6.grid(axis='y', alpha=0.75)
ax6.set_xlim(right=250)  
ax6.set_ylim(top=160)

# VelocitÃ 
ax7 = plt.subplot2grid((3, 3), (2, 2))
ax7.hist(df['Speed Stat'], bins=20, color='gold', edgecolor='black')
ax7.set_title("VelocitÃ ")
ax7.set_xlabel('VelocitÃ ')
ax7.set_ylabel('Frequenza')
ax7.grid(axis='y', alpha=0.75)
ax7.set_xlim(right=250)  
ax7.set_ylim(top=160)

plt.tight_layout()
plt.show()
```
